```python
import pandas as pd
import sqlite3
```

```python
df=pd.read_csv("/content/Dataset for Data Analytics - Sheet1.csv")
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("Duplicate rows:", df.duplicated().sum())

df.drop_duplicates(inplace=True)

print("After removing duplicates:", df.shape)
```

```python
df.columns = df.columns.str.strip()
```

```python
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].astype(str).str.strip()
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
```

```python
df["CouponCode"] = df["CouponCode"].replace("nan", "No Coupon")
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
conn = sqlite3.connect("sales_analysis.db")
```

```python
df.to_sql("orders", conn, if_exists="replace", index=False)

print("Data inserted into SQL database successfully.")
```

```python
def run_query(query):
    return pd.read_sql_query(query, conn)
```

```python
query = """
SELECT *
FROM orders
LIMIT 10;
"""

run_query(query)
```

```python
query = """
SELECT COUNT(*) AS total_orders
FROM orders;
"""

run_query(query)
```

```python
query = """
SELECT SUM(TotalPrice) AS total_revenue
FROM orders;
"""

run_query(query)
```

```python
query = """
SELECT AVG(TotalPrice) AS average_order_value
FROM orders;
"""

run_query(query)
```

```python
query = """
SELECT OrderID, Product, Quantity, UnitPrice, TotalPrice, OrderStatus
FROM orders
WHERE OrderStatus = 'Delivered';
"""

run_query(query)
```

```python
query = """
SELECT OrderID, CustomerID, Product, Quantity, TotalPrice
FROM orders
WHERE TotalPrice > 1000
ORDER BY TotalPrice DESC;
"""

run_query(query)
```

```python
query = """
SELECT
    Product,
    COUNT(OrderID) AS total_orders,
    SUM(Quantity) AS total_quantity_sold,
    SUM(TotalPrice) AS total_revenue,
    AVG(TotalPrice) AS average_revenue
FROM orders
GROUP BY Product
ORDER BY total_revenue DESC;
"""

run_query(query)
```

```python
query = """
SELECT
    OrderStatus,
    COUNT(OrderID) AS total_orders,
    SUM(TotalPrice) AS total_revenue
FROM orders
GROUP BY OrderStatus
ORDER BY total_orders DESC;
"""

run_query(query)
```

```python
query = """
SELECT
    ReferralSource,
    COUNT(OrderID) AS total_orders,
    SUM(TotalPrice) AS total_revenue
FROM orders
GROUP BY ReferralSource
ORDER BY total_revenue DESC;
"""

run_query(query)
```

```python
query = """
SELECT
    CouponCode,
    COUNT(OrderID) AS total_orders,
    SUM(TotalPrice) AS total_revenue
FROM orders
GROUP BY CouponCode
ORDER BY total_orders DESC;
"""

run_query(query)
```

```python
query = """
SELECT
    OrderID,
    CustomerID,
    Product,
    Quantity,
    UnitPrice,
    TotalPrice
FROM orders
ORDER BY TotalPrice DESC
LIMIT 10;
"""

run_query(query)
```

```python
query = """
SELECT
    Product,
    SUM(Quantity) AS total_quantity,
    AVG(Quantity) AS average_quantity,
    MIN(Quantity) AS minimum_quantity,
    MAX(Quantity) AS maximum_quantity
FROM orders
GROUP BY Product
ORDER BY total_quantity DESC;
"""

run_query(query)
```

```python
query = """
SELECT
    strftime('%Y-%m', Date) AS month,
    COUNT(OrderID) AS total_orders,
    SUM(TotalPrice) AS monthly_revenue
FROM orders
GROUP BY month
ORDER BY month;
"""

run_query(query)
```

```python
query = """
SELECT
    CustomerID,
    COUNT(OrderID) AS total_orders,
    SUM(TotalPrice) AS total_spent,
    AVG(TotalPrice) AS average_spent
FROM orders
GROUP BY CustomerID
ORDER BY total_spent DESC
LIMIT 10;
"""

run_query(query)
```

```python
# ==============================
# Step 5: Save Important SQL Results
# ==============================

product_sales = run_query("""
SELECT
    Product,
    COUNT(OrderID) AS total_orders,
    SUM(Quantity) AS total_quantity_sold,
    SUM(TotalPrice) AS total_revenue,
    AVG(TotalPrice) AS average_revenue
FROM orders
GROUP BY Product
ORDER BY total_revenue DESC;
""")

payment_analysis = run_query("""
SELECT
    PaymentMethod,
    COUNT(OrderID) AS total_orders,
    SUM(TotalPrice) AS total_revenue
FROM orders
GROUP BY PaymentMethod
ORDER BY total_revenue DESC;
""")

status_analysis = run_query("""
SELECT
    OrderStatus,
    COUNT(OrderID) AS total_orders,
    SUM(TotalPrice) AS total_revenue
FROM orders
GROUP BY OrderStatus
ORDER BY total_orders DESC;
""")

monthly_revenue = run_query("""
SELECT
    strftime('%Y-%m', Date) AS month,
    COUNT(OrderID) AS total_orders,
    SUM(TotalPrice) AS monthly_revenue
FROM orders
GROUP BY month
ORDER BY month;
""")

product_sales.to_csv("product_sales_summary.csv", index=False)
payment_analysis.to_csv("payment_method_analysis.csv", index=False)
status_analysis.to_csv("order_status_analysis.csv", index=False)
monthly_revenue.to_csv("monthly_revenue_analysis.csv", index=False)

print("SQL output files saved successfully.")
```
